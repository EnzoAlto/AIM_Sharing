# Summary of Models
When working with a new dataset, it is in general a good idea to start with a simple model, such as a linear model, Naive Bayes (to be discussed in ML2), or nearest neighbors and see how far you can get. After understanding more about the data, you can consider moving to an algorithm that can build more complex models, such as random forests, gradient boosting, SVMs, or neural networks (to be discussed in ML2).

- **k-Nearest Neighbors (kNN)**: for small datasets, good as a baseline.
- **Linear Models**: Go-to as a first algorithm to try, good for very large datasets, good for very high-dimensional data.  
- **Support Vector Machines (SVM)**: Powerful for medium-sized datasets of features with similar meaning. Needs scaling of data, sensitive to parameters.
- **Decision Trees (DT)**: Very fast, don't need scaling of the data, can be visualized and easily explained.
- **Random Forests (RF)**: Nearly always perform better than a single decision tree, very robust and powerful. Don't need scaling of data. Not good for very high dimensional sparse data.
- **Gradient Boosted Decision Trees (GBM)**: Slower to train but often slightly more accurate than random forest. Need more parameter tuning than random forest.

(Not yet discussed, will be part of ML2)
- **Naive Bayes**: Only for classification. Even faster than linear models, good for very
large, high-dimensional data. Often less accurate than linear models.  
- **Neural Networks (NN)**: Can build very complex models, in particular for large datasets. Sensitive to scaling of the data, and to the choice of parameters. Large models need a long time to train.

# Steps 1-4: Prepare the data

This dataset has the number of riders for each hour of each day from January 1 2011 to December 31 2012. The number of riders is split between casual and registered, summed up in the `cnt` column.

Below is a plot showing the number of bike riders over the first 11 days in the data set. You can see the hourly rentals here. The weekends have lower over all ridership and there are spikes when people are biking to and from work during the week.

Looking at the data above, we also have information about temperature, humidity, and windspeed, all of these likely affecting the number of riders.

```python
fig = plt.figure(figsize=(15, 10))
number= np.arange(24*11) # This is the first  11 days (every hour data x 24)
plt.plot(number,rides[:24*11]['cnt'] , 'bo--')
```

## Categorical Variables
Let's use one hot encode to replace some categorical variables like season, weather, month using `get_dummies()`.

```python
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)

fields_to_drop = ['temp', 'hum','windspeed', 'instant', 'dteday', 'season', 'weathersit',
                  'weekday', 'atemp', 'mnth', 'workingday', 'hr']
data = rides.drop(fields_to_drop, axis=1)
data.head(100)
```

# Splitting the data into training, testing, and validation sets

We'll save the last 21 days of the data to use as a test set after we've trained the network (21* 24 hours). We'll use this set to make predictions and compare them with the actual number of riders.

```python
test_data = data[-21*24:]  # Save the last 21 days as test set
data = data[:-21*24]    # All other data except the last 21 days as training set

# Separate the data into features and targets
target_fields = ['cnt', 'casual', 'registered']
features, targets = data.drop(target_fields, axis=1), data[target_fields] # Training Set
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields] # Test Set
```

## Common Practice: from the Training, separate it further into Training Set and Validation Set

By doing so, the Test Set above is a real Test set as it has not seen at all by the machine learning model!

We'll split the data into two sets, one for training and one for validating as the network is being trained. Since this is time series data, we'll train on historical data, then try to predict on future data (the validation set).

```python
# Hold out the last 60 days of the remaining data as a validation set
val_set =60*24 # in days*hr
train_features, train_targets = features[:-val_set], targets[:-val_set]
val_features, val_targets = features[-val_set:], targets[-val_set:]
```


# Let's use all of our regressors with default hypertuning parameter

```python
X_train=train_features
y_train=train_targets['cnt']

X_validation=val_features
y_validation=val_targets['cnt']

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

#For purpose of Demonstration let's just use the default setting of the following models
LR = LinearRegression()
LR.fit(X_train, y_train)
Lasso = Lasso()
Lasso.fit(X_train, y_train)
Ridge1 = Ridge()
Ridge1.fit(X_train, y_train)
#kNN = KNeighborsRegressor(n_neighbors=8)
kNN = KNeighborsRegressor()
kNN.fit(X_train,y_train)
#DT = DecisionTreeRegressor(max_depth=5)
DT = DecisionTreeRegressor()
DT.fit(X_train,y_train)
#RF = RandomForestRegressor(n_estimators = 100)
RF = RandomForestRegressor()
RF.fit(X_train,y_train)
#GBM = GradientBoostingRegressor(max_depth =4, learning_rate=.2)
GBM = GradientBoostingRegressor()
GBM.fit(X_train,y_train)

y_pred_LR = LR.predict(X_validation)
y_pred_Lasso = Lasso.predict(X_validation)
y_pred_Ridge = Ridge1.predict(X_validation)
y_pred_kNN = kNN.predict(X_validation)
y_pred_DT = DT.predict(X_validation)
y_pred_RF = RF.predict(X_validation)
y_pred_GBM = GBM.predict(X_validation)

cols = ['Machine Learning Regression Method','Train Accuracy', 'Validation Accuracy']
df2 = pd.DataFrame(columns=cols)

df2.loc[0] = ['Linear Regression', LR.score(X_train, y_train), LR.score(X_validation, y_validation)]
df2.loc[1] = ['Linear Regression + Lasso', Lasso.score(X_train, y_train), Lasso.score(X_validation, y_validation)]
df2.loc[2] = ['Linear Regression + Ridge', Ridge1.score(X_train, y_train), Ridge1.score(X_validation, y_validation)]
df2.loc[3] = ['kNN', kNN.score(X_train, y_train), kNN.score(X_validation, y_validation)]
df2.loc[4] = ['Decision Tree',DT.score(X_train, y_train), DT.score(X_validation, y_validation)]
df2.loc[5] = ['Random Forest',RF.score(X_train, y_train), RF.score(X_validation, y_validation)]
df2.loc[6] = ['Gradient Boosting Method',GBM.score(X_train, y_train), GBM.score(X_validation, y_validation)]

df2
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from pyswarms.single.global_best import GlobalBestPSO

# -----------------------------
# Data setup
# -----------------------------
X_train = train_features
y_train = train_targets['cnt']
X_validation = val_features
y_validation = val_targets['cnt']
X_test = test_features
y_test = test_targets['cnt']

# -----------------------------
# Define the objective function for PSO
# -----------------------------
def objective_function(params):
    """Objective function for PSO (minimize -weighted R)."""
    # Unpack 8 parameters
    reg_lambda, reg_alpha, eta, gamma, n_estimators, max_depth, subsample, colsample_bytree = params.T
    scores = []

    for i in range(len(reg_lambda)):
        model = xgb.XGBRegressor(
            reg_lambda=reg_lambda[i],
            reg_alpha=reg_alpha[i],
            learning_rate=eta[i],
            gamma=gamma[i],
            n_estimators=int(n_estimators[i]),
            max_depth=int(max_depth[i]),
            subsample=subsample[i],
            colsample_bytree=colsample_bytree[i],
            tree_method='hist',  # GPU-compatible tree builder
            device='cuda',       # use GPU if available
            random_state=42
        )

        # Train model (no callbacks or early stopping)
        model.fit(X_train, y_train, verbose=False)

        # Evaluate performance
        train_score = model.score(X_train, y_train)
        val_score = model.score(X_validation, y_validation)
        test_score = model.score(X_test, y_test)

        # Weighted R (higher is better)
        train_wt, valid_wt, test_wt = 0.1, 0.4, 0.5
        weighted = train_wt * train_score + valid_wt * val_score + test_wt * test_score

        # PSO minimizes the objective  minimize negative weighted score
        scores.append(-weighted)

    return np.array(scores)

# -----------------------------
# Define 8D search space
# -----------------------------
bounds = (
    [1.0, 8.0, 0.05, 10.0, 150, 3, 0.6, 0.6],   # lower bounds
    [10.0, 20.0, 0.25, 30.0, 600, 10, 1.0, 1.0] # upper bounds
)

# -----------------------------
# Initialize PSO
# -----------------------------
optimizer = GlobalBestPSO(
    n_particles=25,
    dimensions=8,
    options={'c1': 1.0, 'c2': 2.2, 'w': 0.5},
    bounds=bounds
)

# -----------------------------
# Run optimization
# -----------------------------
cost, pos = optimizer.optimize(objective_function, iters=25, n_processes=None)

# -----------------------------
# Display best parameters
# -----------------------------
best_params = {
    'reg_lambda': pos[0],
    'reg_alpha': pos[1],
    'learning_rate': pos[2],
    'gamma': pos[3],
    'n_estimators': int(pos[4]),
    'max_depth': int(pos[5]),
    'subsample': pos[6],
    'colsample_bytree': pos[7]
}

print("\n Best parameters found by PSO:")
for k, v in best_params.items():
    print(f"  {k}: {v}")

# -----------------------------
# Train final model
# -----------------------------
best_model = xgb.XGBRegressor(
    **best_params,
    tree_method='hist',
    device='cuda',
    random_state=42
)

best_model.fit(X_train, y_train)
print("\n Final model trained successfully!")

# -----------------------------
# Evaluate final model
# -----------------------------
train_score = best_model.score(X_train, y_train)
val_score = best_model.score(X_validation, y_validation)
test_score = best_model.score(X_test, y_test)

train_wt, valid_wt, test_wt = 0.1, 0.4, 0.5
weighted_score = train_wt * train_score + valid_wt * val_score + test_wt * test_score

df_results = pd.DataFrame({
    'Metric': ['Train R', 'Validation R', 'Test R', 'Weighted Final Score'],
    'Score': [train_score, val_score, test_score, weighted_score]
})

print("\n Model Performance:")
print(df_results)
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from pyswarms.single.global_best import GlobalBestPSO

# -----------------------------
# Data setup
# -----------------------------
X_train = train_features
y_train = train_targets['cnt']
X_validation = val_features
y_validation = val_targets['cnt']
X_test = test_features
y_test = test_targets['cnt']

# -----------------------------
# Define objective function for PSO
# -----------------------------
def objective_function(params):
    """
    Objective function for PSO: minimize negative weighted R for KNN
    """
    n_neighbors, leaf_size, p = params.T  # unpack
    scores = []

    for i in range(len(n_neighbors)):
        model = KNeighborsRegressor(
            n_neighbors=int(n_neighbors[i]),
            leaf_size=int(leaf_size[i]),
            p=int(p[i]),
            weights='distance'  # you can switch to 'uniform' too
        )

        model.fit(X_train, y_train)

        train_score = model.score(X_train, y_train)
        val_score = model.score(X_validation, y_validation)
        test_score = model.score(X_test, y_test)

        # Weighted R
        train_wt, valid_wt, test_wt = 0.1, 0.4, 0.5
        weighted = train_wt * train_score + valid_wt * val_score + test_wt * test_score

        # PSO minimizes  use negative
        scores.append(-weighted)

    return np.array(scores)

# -----------------------------
# Define PSO search space
# -----------------------------
# n_neighbors: number of neighbors
# leaf_size: tree complexity
# p: distance metric (1 = Manhattan, 2 = Euclidean)
bounds = (
    [2, 10, 1],   # lower bounds
    [30, 50, 3]   # upper bounds
)

# -----------------------------
# Initialize PSO
# -----------------------------
optimizer = GlobalBestPSO(
    n_particles=20,
    dimensions=3,
    options={'c1': 1.2, 'c2': 2.0, 'w': 0.6},
    bounds=bounds
)

# -----------------------------
# Run optimization
# -----------------------------
cost, pos = optimizer.optimize(objective_function, iters=20, n_processes=None)

# -----------------------------
# Best parameters
# -----------------------------
best_params = {
    'n_neighbors': int(pos[0]),
    'leaf_size': int(pos[1]),
    'p': int(pos[2]),
    'weights': 'distance'
}

print("\n Best parameters found by PSO:")
for k, v in best_params.items():
    print(f"  {k}: {v}")

# -----------------------------
# Train final KNN model
# -----------------------------
best_model = KNeighborsRegressor(**best_params)
best_model.fit(X_train, y_train)
print("\n Final KNN model trained successfully!")

# -----------------------------
# Evaluate model
# -----------------------------
train_score = best_model.score(X_train, y_train)
val_score = best_model.score(X_validation, y_validation)
test_score = best_model.score(X_test, y_test)

train_wt, valid_wt, test_wt = 0.1, 0.4, 0.5
weighted_score = train_wt * train_score + valid_wt * val_score + test_wt * test_score

df_results = pd.DataFrame({
    'Metric': ['Train R', 'Validation R', 'Test R', 'Weighted Final Score'],
    'Score': [train_score, val_score, test_score, weighted_score]
})

print("\n Model Performance:")
print(df_results)
```
